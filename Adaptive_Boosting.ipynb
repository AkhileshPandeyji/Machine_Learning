```python
import pandas as pd
import numpy as np
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
mush_dataset = pd.read_csv('mushrooms.csv')
```

```python
mush_dataset.head()
```

```python
col = list(mush_dataset.columns.values)
```

```python
print(col)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for item in col:
    mush_dataset[item] = LabelEncoder().fit_transform(mush_dataset[item])
```

```python
print(mush_dataset.shape)
```

```python
print(mush_dataset.info())
```

```python
mush_dataset.iloc[:100,:]
```

```python
decision_tree = DecisionTreeClassifier(criterion='entropy',max_depth=10)
```

```python
ada_model = AdaBoostClassifier(base_estimator=decision_tree,n_estimators=100,learning_rate=1.0)
```

```python
y = mush_dataset.iloc[:,0:1].values
X = mush_dataset.iloc[:,1:].values
```

```python
print(len(y[0]),len(X[0]))
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
```

```python
ada_model = ada_model.fit(X_train,y_train)
```

```python
y_pred = ada_model.predict(X_test)
```

```python
accuracy = accuracy_score(y_pred,y_test)
```

```python
print(accuracy*100)
```

```python
for i in range(10):
    print('Actual : {} , Predicted : {}.'.format(y_pred[i],y_test[i]))
```
