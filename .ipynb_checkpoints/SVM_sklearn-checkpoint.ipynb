```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
dataset = pd.read_csv("svm_sklearn_dataset.csv")
```

```python
print(dataset.shape)
```

```python
print(dataset.head())
```

```python
X = dataset.iloc[:,2:4].values
y = dataset.iloc[:,-1].values
```

```python
sc = StandardScaler()
```

```python
X = sc.fit_transform(X)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
svm = SVC()
```

```python
svm.fit(X_train,y_train)
```

```python
y_pred = svm.predict(X_test)
```

```python
print(accuracy_score(y_pred,y_test))
```

```python
plt.scatter(X_train[:,0],X_train[:,1],c=y_train)
```

```python
from matplotlib.colors import ListedColormap
```

```python
xx = np.arange(X_train[:,0].min()-0.5,X_train[:,0].max()+0.5,0.01)
yy = np.arange(X_train[:,1].min()-0.5,X_train[:,1].max()+0.5,0.01)
```

```python
x1,x2 = np.meshgrid(xx,yy)
```

```python
print(x1.shape,x2.shape)
```

```python
z = svm.predict(np.asarray([x1.ravel(),x2.ravel()]).T).reshape(x1.shape)
```

```python
plt.contourf(x1,x2,z,cmap = ListedColormap(('cyan','b')))
plt.scatter(X_train[:,0],X_train[:,1],c=y_train)
```
