```python
from sklearn.datasets import make_moons
```

```python
features,target = make_moons(n_samples=1000,noise=0.1)
```

```python
print(features)
```

```python
import matplotlib.pyplot as plt
plt.scatter(features[:,0],features[:,1],c=target)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='rbf')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(features,target,test_size=0.2)
```

```python
svc.fit(x_train,y_train)
```

```python
ypred = svc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(ypred,y_test))
```

# SVM - meshgrid analysis

```python
from matplotlib.colors import ListedColormap
import numpy as np
```

```python
xx = np.arange(x_train[:,0].min()-0.5,x_train[:,0].max()+0.5,0.02)
yy = np.arange(x_train[:,1].min()-0.5,x_train[:,1].max()+0.5,0.02)
```

```python
x1,x2 = np.meshgrid(xx,yy)
```

```python
print(x1.shape,x2.shape)
```

```python
z =  svc.predict(np.asarray([x1.ravel(),x2.ravel()]).T).reshape(x1.shape)
```

```python
plt.contourf(x1,x2,z,cmap = ListedColormap(('cyan','red')))
plt.scatter(x_train[:,0],x_train[:,1],c=y_train)
```
