```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
style.use('fivethirtyeight')
```

```python
matches = pd.read_csv('matches.csv')
deliveries = pd.read_csv('deliveries.csv')
```

```python
matches['season'].value_counts()
```

```python
#Matches per season
plt.figure(figsize=(10,6))
plt.title('Matches Played in Every Season')
sns.countplot(x='season',data=matches)
```

```python
#How toss decisions varied over seasons
plt.figure(figsize=(10,6))
sns.countplot(x='season',hue='toss_decision',data=matches)
```

```python
#Pie chart for the teams that won the toss and won the match
temp_df = matches[matches['toss_winner'] == matches['winner']]
```

```python
temp_df
```

```python
# yes or no counts array
slices = [len(temp_df),len(matches)-len(temp_df)]
```

```python
slices
```

```python
#pie chart plotting
plt.figure(figsize=(10,6))
plt.pie(slices,labels=['Yes','No'],colors=['r','g'],autopct='%1.2f%%',shadow=True,explode=(0.06,0),startangle=90)
plt.legend(loc=0)
plt.show()
```

```python
#most number of toss winners
x= matches['toss_winner'].value_counts()
```

```python
plt.figure(figsize=(10,6))
ax = x.plot.bar(width=0.9,color=sns.color_palette('inferno'))

for p in ax.patches:
    ax.annotate(format(p.get_height()),(p.get_x()+0.3,p.get_height()+0.5))
```

```python
#most sixes
six_data = deliveries[deliveries['batsman_runs'] == 6]
y = six_data['batsman'].value_counts()[:10]
```

```python
plt.figure(figsize=(10,6))
ax = y.plot.bar(width=0.8,color=sns.color_palette('RdYlGn'))

for p in ax.patches:
    ax.annotate(format(p.get_height()),(p.get_x()+0.3,p.get_height()+4))    
```

```python
#dataframe object function groupby
#first 10 most runs scoring batsmans
most_runs = deliveries.groupby('batsman')['batsman_runs'].sum()
```

```python
most_runs = most_runs.sort_values(ascending=False)[0:10]
most_runs
```

```python
plt.figure(figsize=(10,6))
ax =most_runs.plot.bar(width=0.8,color=sns.color_palette('inferno'))
for p in ax.patches:
    ax.annotate(format(p.get_height()),(p.get_x()+0.3,p.get_height()+25))
```

```python
#most runs scored by batsman in an innings - left se hoti hai grouping aur match_id is to the left of batsman
mi_runs = deliveries.groupby(('match_id','batsman'))['batsman_runs'].sum().sort_values(ascending=False)[0:10]
```
