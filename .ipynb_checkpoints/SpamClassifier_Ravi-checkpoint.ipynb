```python
import io,os
import email
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer,TfidfTransformer
from sklearn.model_selection import train_test_split
```

```python
def read_path(path):
    for root,folder,files in os.walk(path):
        for file in files:
            filename = root + '\\' + file
            f = io.open(filename,encoding='latin1')
            lines = f.readlines()
            data = []
            body = False
            for line in lines:
                if line == '\n':
                    body = True
                if body == True:
                    data.append(line)
            msg = '\n'.join(data)
            yield msg,filename
```

```python
def df(path,classification):
    index = []
    row = []
    for message,filename in read_path(path):
        row.append({'message':message,'classification':classification})
        index.append(filename)
    return pd.DataFrame(data=row,index=index)
```

```python
dataset = pd.DataFrame()
dataset = df('emails/ham','ham')
dataset.append(df('emails/spam','spam'))
```

```python
import re
def textCleaning(arr):
    temp = re.findall('[\w]+',arr)
    clean = ' '.join(temp)
    return clean
```

```python
for i in range(len(dataset['message'])):
    dataset['message'][i] = textCleaning(dataset['message'][i])
```
