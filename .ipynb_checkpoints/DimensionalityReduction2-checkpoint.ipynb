```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
def embedding_plot(X, title):
    x_min, x_max = np.min(X, axis=0), np.max(X, axis=0)
    X = (X - x_min) / (x_max - x_min)
 
    plt.figure(figsize=(20,10))
    ax = plt.subplot(aspect='equal')
    sc = ax.scatter(X[:,0], X[:,1], lw=0, s=40, c=y/10.)
 
    shown_images = np.array([[1., 1.]])
    for i in range(X.shape[0]):
        if np.min(np.sum((X[i] - shown_images) ** 2, axis=1)) < 1e-2: continue
        shown_images = np.r_[shown_images, [X[i]]]
        ax.add_artist(offsetbox.AnnotationBbox(offsetbox.OffsetImage(digits.images[i], cmap=plt.cm.gray_r), X[i]))
 
    plt.xticks([]), plt.yticks([])
    plt.title(title)
```

```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
X = digits.data
```

```python
y = digits.target
```

```python
print(X[:5])
```

```python
print(y[:5])
```

```python
n_samples,n_features = X.shape
```

```python
from sklearn.decomposition import PCA
from matplotlib import offsetbox
```

```python
X_pca = PCA(n_components=2).fit_transform(X)
embedding_plot(X_pca,'PCA')
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
X_lda = LinearDiscriminantAnalysis(n_components=2).fit_transform(X,y)
```

```python
embedding_plot(X_lda,"LinearDiscriminantAnalysis")
```

```python
from sklearn.manifold import TSNE
```

```python
X_tsne = TSNE(n_components=2,init='pca').fit_transform(X)
embedding_plot(X_tsne,"T-Distributed Stochastic Neighbor Embedding")
```
